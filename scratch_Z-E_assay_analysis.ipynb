```python
df_ = pd.read_csv(file, sep=';')
df_ = df_.fillna(0)
df_.replace(0, np.nan, inplace=True)
df_
```

```python
file = path.joinpath('cell_line_profile.csv')
df_prof = pd.read_csv(file, sep=';')
df_prof = df_prof.drop('ID / plate', axis=1)
gr = ['cell line', 'type']
df_prof = df_prof.groupby(gr).mean() * 2.18  # (um)
df_prof = df_prof.loc[df_[gr[0]].unique(),df_[gr[1]].unique(),:] 
df_prof
```

```python
df = pd.merge(df_, df_prof, on=gr, how='right')
profile = df[df_prof.columns]
```

```python
labels = ['type', 'cell line', 'investigation'] 
x = df.drop(labels + ['cell size (um)', 'Látóterek'] + list(profile.columns), axis=1)
y_ = df[labels]
y_.drop_duplicates()
```

```python
x
```

```python
f = x.copy()
dcols = x.columns

f['mean'] = x.mean(axis=1)
f['std'] = x.std(axis=1)
f['kurtosis'] = x.kurtosis(axis=1)
f['skew'] = x.skew(axis=1)

shrink = x.diff(axis=1) * -1
# display(shrink)
f['shrink mean'] = shrink.mean(axis=1)
f['shrink std'] = shrink.std(axis=1)
f['shrink kurtosis'] = shrink.kurtosis(axis=1)
f['shrink skew'] = shrink.skew(axis=1)

# normed with cell profile data
for tp in ['', 'shrink ']:
    for col in ['mean', 'std', 'kurtosis', 'skew']:
        for cell_par in profile.columns:
            f[f'{cell_par}_norm: {tp}{col}'] = f[f'{tp}{col}'] / profile[cell_par] * 2.18  # (um)
            # break;
        # f[f'ContArea_norm: {tp}{col}'] = f[f'{tp}{col}'] / df_['cell size (um)'] * 2.18  # (um)

def get_merged_state(df):
    m_list = []
    df = df.drop('0h', axis=1)
    cols = df.columns
    for ind in df.index:
        merged = False
        for i, el in enumerate(df.loc[ind]):
            if not (el >= 0):
                m_list.append(int(cols[i].strip('h')))
                merged = True
                break
        if not merged:
            m_list.append(np.nan)
    return m_list
        

f['merged'] = get_merged_state(shrink)
# f[f['merged'].isnull()]
f.drop(x.columns, axis=1)
```

```python
no_profile_mask = ['_norm:' not in  c for c in f.columns]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedShuffleSplit

# X = x
# X = f.loc[:, no_profile_mask]
X = f
# X = f.drop(x.columns, axis=1)
X = X.fillna(-1)

VERBOSE = False

REPEAT_CV = 10
selection = 15  # first n feature

# cv = StratifiedKFold(n_splits=5)
cv = StratifiedShuffleSplit(n_splits=5)
pipe = RandomForestClassifier(n_estimators=500, n_jobs=-2)
# pipe = make_pipeline(
#         # PCA(n_components=.97),
#         StandardScaler(),
#         RandomForestClassifier(n_estimators=500, n_jobs=-2)
# )
# pipe = get_ensemble_clf('voting', 'hard')

# print(X.isnull().values.any())
# with pd.option_context('display.max_rows', None):
#     display(X)
X
```

```python
from utils import make_cv, get_importance

ver=VERBOSE
print(f'Investigate poison: {y_["type"].unique()}\n')

for method in y_['investigation'].unique():
    met_ind = y_['investigation'] == method
    mx, my = X.loc[met_ind], y_.loc[met_ind]
    imp = get_importance(mx, my, sort=True)
    most_imp = imp[:selection]
    feat = most_imp.index

    print(f'Method: {method} - Selected {len(feat)} features:\n{list(feat)}\n')
    data = mx[feat]

    for cell_line in my['cell line'].unique():
        if ver:
            print(f'\n======================================================')
        print(f'Investigation: {cell_line}')

        ind = my['cell line'] == cell_line

        make_cv(pipe, cv, data.loc[ind].values, my.loc[ind]['type'].values, repeat=REPEAT_CV, verbose=ver)
    # print()

    print(f'\n{method} - Controlls: {my["cell line"].unique()}\n')
    ind = my['type'] == 'kontroll'
    make_cv(pipe, cv, data.loc[ind].values, my.loc[ind]['cell line'].values, repeat=REPEAT_CV, verbose=ver)
```

```python
ver=VERBOSE
print(f'Investigate Controlls: {y_["cell line"].unique()} - all features')
for method in y_['investigation'].unique():
    if ver:
        print(f'\n======================================================')
    print(f'Investigation: {method}')

    ind = (y_['investigation'] == method) & (y_['type'] == 'kontroll')
    
    make_cv(pipe, cv, X.loc[ind].values, y_.loc[ind]['cell line'].values, repeat=REPEAT_CV, verbose=ver)
```
